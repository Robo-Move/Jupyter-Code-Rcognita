```python
file_path1 = "".join([file_path, file_pathA, file_name1])
file_path2 = "".join([file_path, file_pathA, file_name2])
file_path3 = "".join([file_path, file_pathA, file_name3])
file_path4 = "".join([file_path, file_pathA, file_name4])
file_path5 = "".join([file_path, file_pathA, file_name5])

df1 = pd.read_csv(file_path1, skiprows=15)
df2 = pd.read_csv(file_path2, skiprows=15)
df3 = pd.read_csv(file_path3, skiprows=15)
df4 = pd.read_csv(file_path4, skiprows=15)
df5 = pd.read_csv(file_path5, skiprows=15)
```

```python
# Plot settings
fig = plt.figure(figsize=(14, 10))

# Plot Trajectory (top-left)
plt.subplot(2, 2, 1)
plt.plot(df1['x [m]'], df1['y [m]'], 'b-', lw=0.8, label='Kbeta=-0.5')
plt.plot(df2['x [m]'], df2['y [m]'], 'g-', lw=0.8, label='Kbeta=-1.0')
plt.plot(df3['x [m]'], df3['y [m]'], 'r-', lw=0.8, label='Kbeta=-2.0')
plt.plot(df4['x [m]'], df4['y [m]'], 'c-', lw=0.8, label='Kbeta=3.0')
plt.plot(df5['x [m]'], df5['y [m]'], 'm-', lw=0.8, label='Kbeta=-4.0')
plt.scatter(df1['x [m]'].iloc[0], df1['y [m]'].iloc[0], marker='o', color='g', lw=0.8, label='Start')
plt.scatter(0, 0, marker='x', color='r', label='End')
plt.xlabel('x [m]')
plt.ylabel('y [m]')
plt.title('Trajectory')
plt.legend()
plt.grid()
plt.xlim(-4, 4)
plt.ylim(-4, 4)

plt.tight_layout()
plt.show()
```
